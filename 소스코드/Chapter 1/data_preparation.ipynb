```python
# 피벗 테이블 제작하기 1 (단일 인덱스를 사용할 시)
df = pd.read_csv('penguins.csv')

df.pivot_table(
    index='species',
    columns='sex',
    values='body_mass_g', aggfunc='mean'
)
```

```python
# 피벗 테이블 만들기 2 (여러 개의 인덱스를 사용할 시)
df.pivot_table(
    index=['species','island'],
    columns='sex',
    values='body_mass_g', aggfunc='mean'
)
```

```python
# 주피터랩에서 작업 중인 pandas 데이터 테이블을 클립보드로 내보내기
df.pivot_table(
    index=['species','island'],
    columns='sex',
    values='body_mass_g', aggfunc='mean'
).to_clipboard()
```

```python
# groupby 메서드를 이용한 그룹별 집계연산
df.groupby('species')['body_mass_g'].max()
```

```python
# groupby 메서드에 사용자 정의 함수 적용
def custom_function(group):
    squared_mean = (group['body_mass_g'] ** 2).mean()
    return squared_mean

df.groupby('species').apply(custom_function)
```
